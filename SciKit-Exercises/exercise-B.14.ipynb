```python
# Exercise B14

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates kde plot of sepal length versus sepal width for Setosa
sub=iris[iris['species']=='Iris-setosa']
sns.kdeplot(data=sub[['sepal_length','sepal_width']], shade=True, thresh=0.05)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
