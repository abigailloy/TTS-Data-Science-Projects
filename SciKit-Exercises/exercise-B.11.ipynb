```python
# Exercise B11

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates scatterplot w/ joint density estimate to describe individual distributions on the same plot b/t Sepal length and Sepal width
sns.jointplot("sepal_length", "sepal_width", data=iris, color="c").plot_joint(sns.kdeplot, zorder=0, n_levels=6) 
plt.show()
```
