```python
# Exercise B12

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates joinplot using "kde" to describe individual distributions on the same plot between Sepal length and Sepal width 
#w/ '+' as a marker
fig = sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind="kde", color="c")
fig.plot_joint(plt.scatter, c="k", s=40, linewidth=0.5, marker="+")
fig.ax_joint.collections[0].set_alpha(0)
fig.set_axis_labels("Sepal Length", "Sepal Width") 
plt.show()
```
