```python
# Exercise B10

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates joinplot w/ regression and kernel density fits using "reg" to describe individual distributions on the same plot b/t Sepal length and Sepal width
fig=sns.jointplot(x='sepal_length', y='sepal_width', kind="reg", color='blue', data=iris) 
plt.show()
```
