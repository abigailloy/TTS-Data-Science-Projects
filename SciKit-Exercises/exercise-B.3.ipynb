```python
# Exercise B3

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#counts # of occurences of each species and creates a pie plot
iris['species'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(5,4))
plt.title("% of Species Occurence in Iris Data")
plt.show()
```
