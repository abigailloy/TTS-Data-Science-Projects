```python
# Exercise B4

import pandas as pd
import matplotlib.pyplot as plt

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates scatter plot to plot relationship b/t sepal length and width
fig = iris[iris.species=='Iris-setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='purple', label='Setosa')
iris[iris.species=='Iris-versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='Versicolor',ax=fig)
iris[iris.species=='Iris-virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='yellow', label='Virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Sepal Width")
plt.show()
```
