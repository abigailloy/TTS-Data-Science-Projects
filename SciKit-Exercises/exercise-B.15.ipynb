```python
# Exercise B15

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates kde plot of petal length vs. petal width for Setosa species of flower
sub=iris[iris['species']=='Iris-setosa']
sns.kdeplot(data=sub[['petal_length','petal_width']], shade=True, thresh=0.05)
plt.title('Iris-Setosa Petal Length vs. Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
