```python
# Exercise B5

import pandas as pd
import matplotlib.pyplot as plt

#loads Iris dataset from url link
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

#renames columns
columnNames = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

#creates 'iris' dataframe
iris = pd.read_csv(url, names = columnNames)

#creates scatter plot to plot relationship b/t petal length and width
fig = iris[iris.species=='Iris-setosa'].plot.scatter(x='petal_length',y='petal_width',color='purple', label='Setosa')
iris[iris.species=='Iris-versicolor'].plot.scatter(x='petal_length',y='petal_width',color='green', label='Versicolor',ax=fig)
iris[iris.species=='Iris-virginica'].plot.scatter(x='petal_length',y='petal_width',color='yellow', label='Virginica', ax=fig)
fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title("Petal Length VS Petal Width")
plt.show()
```
